# Create a pipeline using spaCy

```
# Importing Pandas and spaCy modules
import pandas as pd
import spacy
```

```
## Reading 'json' type dataset into pandas dataframe
df = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz", lines=True)
df.head()
```

```
# loading the spaCy core English language model
sp = spacy.load('en_core_web_sm')
```

### For a single sentence/string

```
# Store the string as a simple spaCy document
sentence = sp(df.reviewText[0])     ## taking the first sentence of reviewText column as i/p str
```

```
# Tokenize the given string
for word in sentence:
    print(word.text)
```

```
# Assign Pos tags to the tokens in the given string
for word in sentence:
    print(word.text + '\t\t',  word.pos_)
```

```
# Lemmatize the tokens in the given string
for word in sentence:
    print(word.text + '\t===>',  word.lemma_)
```

### For multiple strings (or entire column)

```
# Function to create a nlp pipeline using spaCy

def spacy_pipeline(str):
  pipe = dict()
  tokens = ' '.join([word.text for word in sp(str)])
  lemmatize = ' '.join([word.lemma_ for word in sp(str)])
  tagged = ' '.join([word.pos_ for word in sp(str)])

  pipe['Tokenize'] = tokens
  pipe['Lemmatize'] = lemmatize
  pipe['Pos_tag'] = tagged

  return pipe
```

```
## Function to find adjectives in the given string
def find_adj(str):
  adj = []
  
  for word in sp(str):
    if word.pos_ == 'ADJ':
      adj.append(word.text)

  return adj
```

```
## Function to find nouns in the given string
def find_noun(str):
  noun = []
  
  for word in sp(str):
    if word.pos_ == 'NOUN':
      noun.append(word.text)
      
  return noun
```

```
df1 = df.iloc[:100]     ## Taking only the first 100 rows of the dataset for input
spacy_pln = df1.reviewText.apply(spacy_pipeline)      ## Creating spaCy pipeline for all rows of ReviewText
df1.insert(5, 'Spacy Pipeline(Dict)_RevTxt', spacy_pln)
df1.head()
```

```
spacy_pln = df1.summary.apply(spacy_pipeline)
df1.insert(8, 'Spacy Pipeline(Dict)_Summary', spacy_pln)  ## Creating spaCy pipeline for all rows of Summary
df1.head()
```

```
adj_rev = df1.reviewText.apply(find_adj)   ## Finding adjectives in each row of ReviewText
df1.insert(6, 'Adjectives_RevTxt', adj_rev)
adj_summ = df1.summary.apply(find_adj)     ## Finding adjectives in each row of Summary
df1.insert(10, 'Adjectives_Summary', adj_summ)
df1.head()
```

```
noun_rev = df1.reviewText.apply(find_noun)      ## Finding nouns in each row of ReviewText
df1.insert(7, 'Nouns_RevTxt', noun_rev)
noun_summ = df1.summary.apply(find_noun)        ## Finding nouns in each row of Summary
df1.insert(12, 'Nouns_Summary', noun_summ)
df1.head()
```
